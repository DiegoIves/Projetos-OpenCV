## Desenho de formas na imagem

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
#Dictionary containing some colors
colors = {'blue': (255, 0, 0), 'green': (0, 255, 0), 'red': (0, 0, 255), 'yellow':
(0, 255, 255), 'magenta': (255, 0, 255), 'cyan': (255, 255, 0), 'white': (255, 255,
255), 'black': (0, 0, 0), 'gray': (125, 125, 125), 'rand': np.random.randint(0,
high=256, size=(3,)).tolist(), 'dark_gray': (50, 50, 50), 'light_gray': (220, 220,
220)}


def show_matplotlib(cor):
    image = np.zeros((500,500,3),dtype="uint8")
    image[:] = cor
    separation = 40
    for key in colors:
        cv2.line(image,(0,separation),(500,separation),colors[key],20)
        separation +=40
        img_matplotlib = image[:,:,::-1]
    
    plt.imshow(img_matplotlib)


show_matplotlib(colors['light_gray'])
```

```python
#Dictionary containing some colors
colors = {'blue': (255, 0, 0), 'green': (0, 255, 0), 'red': (0, 0, 255), 'yellow':
(0, 255, 255), 'magenta': (255, 0, 255), 'cyan': (255, 255, 0), 'white': (255, 255,
255), 'black': (0, 0, 0), 'gray': (125, 125, 125), 'rand': np.random.randint(0,
high=256, size=(3,)).tolist(), 'dark_gray': (50, 50, 50), 'light_gray': (220, 220,
220)}
image = np.zeros((400,400,3),dtype="uint8")
image[:] = colors['light_gray']

def show_matplotlib(img):
    cv2.circle(img,(50,50),20,colors['green'],3)
    cv2.line(img,(50,50),(0,50),colors["red"],1)
    cv2.line(img,(50,50),(50,0),colors["red"],1)

    cv2.circle(img,(100,100),30,colors['blue'],-1)
    cv2.circle(img,(200,200),40,colors['magenta'],10)
    cv2.circle(img,(300,300),40,colors['cyan'],-1)


    img_matplotlib = img[:,:,::-1]
    
    plt.imshow(img_matplotlib)


show_matplotlib(image)
```

### Advanced Shapes

```python
image = np.zeros((300,300,3),dtype="uint8")
image[:] = colors["light_gray"]

def show_matplotlib(img):
    cv2.line(img,(0,0),(300,300),colors['green'],3)
    cv2.rectangle(img,(0,0),(100,100),colors['blue'],3)
    ret,p1,p2 = cv2.clipLine((0,0,100,100),(0,0),(300,300))
    if ret:
        cv2.line(img,p1,p2,colors["cyan"],3)

    img_matplotlib = img[:,:,::-1]
    
    plt.imshow(img_matplotlib)

show_matplotlib(image)
```

```python
image = np.zeros((300,300,3),dtype="uint8")
image[:] = colors["light_gray"]

def show_matplotlib(img):
    cv2.arrowedLine(img,(50,50),(200,50),colors['red'],3,8,0,0.1)
    cv2.arrowedLine(img,(50,120),(200,120),colors['green'],3,cv2.LINE_AA,0,0.3)
    cv2.arrowedLine(img,(50,200),(200,200),colors['blue'],3,8,0,0.3)

    img_matplotlib = img[:,:,::-1]
    
    plt.imshow(img_matplotlib)

show_matplotlib(image)
```

```python
image = np.zeros((300,300,3),dtype="uint8")
image[:] = colors["light_gray"]

def show_matplotlib(img):
    cv2.ellipse(image, (80, 80), (60, 40), 0, 0, 360, colors['red'], -1)
    cv2.ellipse(image, (80, 200), (80, 40), 0, 0, 360, colors['green'], 3)
    cv2.ellipse(image, (80, 200), (10, 40), 0, 0, 360, colors['blue'], -1)
    cv2.ellipse(image, (200, 200), (10, 40), 0, 0, 180, colors['yellow'], 3)
    cv2.ellipse(image, (200, 100), (10, 40), 0, 0, 270, colors['cyan'], 3)
    cv2.ellipse(image, (250, 250), (30, 30), 0, 0, 360, colors['magenta'], 3)
    cv2.ellipse(image, (250, 100), (20, 40), 45, 0, 360, colors['gray'], 3)

    img_matplotlib = img[:,:,::-1]
    
    plt.imshow(img_matplotlib)

show_matplotlib(image)
```
